```python
import edward as ed
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import tensorflow as tf
import mobilenet_v1
CLASS_OF_INTEREST = 10
EPS = 0.1
sess = tf.Session()
```

## Get some pretrained tensorflow network for our needs


$y = f(x)$, where y --- prediction, $f(\cdot)$ -- NN, x -- image(for example)

```python
D = 224 * 224 * 3
theta = tf.Variable(tf.random_uniform([224, 224, 3], minval=0.1, maxval=0.9), name="theta")
mu = [0.] * D
diag_stdev = [.001] * D
p_dist = tf.contrib.distributions.MultivariateNormalDiag(mu, diag_stdev)
p = tf.Variable(tf.reshape(p_dist.sample(), [224, 224, 3]), name="p")
```

```python
theta_old = tf.Variable(tf.zeros([224, 224, 3]), name="theta_old")
p_old = tf.Variable(tf.zeros([224, 224, 3]), name="p_old")
```

```python
#classifier = tf.sigmoid(logits)[:, CLASS_OF_INTEREST][0]
#log_joint = 10 * (classifier - 1 + EPS)
# log_joint = 10000000000 * logits[:, CLASS_OF_INTEREST][0]
dist = tf.contrib.distributions.Normal(tf.zeros_like(theta), tf.ones_like(theta))
log_joint = tf.reduce_mean(tf.log(dist.prob(theta)))
# p(x) ~ exp(10 * (y - 1 + eps))
step = tf.constant(20., name='step')
one = tf.constant(1.)
two = tf.constant(2.)
```

```python
def train_step(theta, p, p_old, theta_old, log_joint, L):
    assign_ops = []
    # calc H at the beginning
    H_old = tf.reduce_sum(log_joint) + tf.reduce_sum(p**2) / two

    # save old
    p_old_assigned_0 = p_old.assign(p); assign_ops.append(p_old_assigned_0)
    theta_old_assigned_0 = theta_old.assign(theta); assign_ops.append(theta_old_assigned_0)

    # draw impulse from normal distribution
    with tf.control_dependencies([H_old, p_old_assigned_0]):
        p_assigned_0 = p.assign(tf.reshape(p_dist.sample(), [224, 224, 3])); assign_ops.append(p_assigned_0)
    
    # integration leap frog step
    with tf.control_dependencies([p_assigned_0]):
        p_assigned_1 = p.assign_add(step * tf.gradients(log_joint, theta)[0] / two); assign_ops.append(p_assigned_1)
    
    for i in range(L):
        with tf.control_dependencies([p_assigned_1, theta_old_assigned_0]):
            assign_ops.append(tf.gradients(log_joint, theta)[0])
            theta_assigned_0 = theta.assign_add(step * p); assign_ops.append(theta_assigned_0)
        if i != L - 1:
            with tf.control_dependencies([theta_assigned_0]):
                p_assigned_1 = p.assign_add(step * tf.gradients(log_joint, theta)[0] / two); assign_ops.append(p_assigned_1)
            
    with tf.control_dependencies([theta_assigned_0]):
        assign_ops.append(tf.gradients(log_joint, theta)[0])
        p_assigned_2 = p.assign_add(step * tf.gradients(log_joint, theta)[0] / two); assign_ops.append(p_assigned_2)
    
    # new H
    with tf.control_dependencies([p_assigned_2]):
        H_new = tf.reduce_sum(log_joint) + tf.reduce_sum(p**2) / two
    
    with tf.control_dependencies([p_old_assigned_0, theta_old_assigned_0, p_assigned_2]):
        alpha = tf.minimum(one, tf.exp(H_new - H_old))
        p_order = tf.random_uniform(shape=[], minval=0., maxval=1., dtype=tf.float32)
        pred = tf.less(p_order, alpha)
        assign_ops.append(p.assign(tf.cond(pred, lambda: p_assigned_2, lambda: p_old_assigned_0)))
        assign_ops.append(theta.assign(tf.cond(pred, lambda: theta_assigned_0, lambda: theta_old_assigned_0)))
    
    return assign_ops
```

```python
train_step_op = train_step(log_joint=log_joint, theta=theta, p=p, p_old=p_old, theta_old=theta_old, L=4)
```

```python
init = tf.global_variables_initializer()
init.run(session=sess)
```

```python
res = sess.run(train_step_op)
```
